```python
# Importamos librerías
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11) # Me da 11 puntos entre 0 y 5
y = x ** 2
```

## Visualizaciones simples

```python
plt.plot(x, y) # Gráfica de la relación x y y
plt.show()
```

```python
plt.plot(x, y, 'ro') # Gráfica roja y con marcador de puntos
```

```python
plt.plot(x, y, 'bs-') # Gráfica azul y con marcador de cuadrados y línea sólida
plt.show()
```

```python
plt.plot(x, y, 'yD:') # Gráfica amarilla y con marcador de diamantes y línea punteada
plt.show()
```

```python
plt.hist(x) # Histograma de x
plt.show()
```

```python
plt.pie(x) # Gráfica de pastel de x
plt.show()
```

```python
plt.scatter(x, y) # Gráfica de dispersión de x y y
plt.show()
```

```python
plt.boxplot(x) # Diagrama de caja de x
plt.show()
```

```python
plt.bar(x, y) # Gráfica de barras de x y y
plt.show()
```

```python
plt.barh(x, y) # Gráfica de barras horizontales de x y y
plt.show()
```

```python
plt.stem(x, y) # Gráfica de tallos y hojas de x y y
plt.show()
```
