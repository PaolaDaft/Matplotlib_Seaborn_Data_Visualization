```python
# Importamos librer√≠as
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11) # Me da 11 puntos entre 0 y 5
y = x ** 2
```

## Visualizaciones simples

```python
plt.plot(x, y) # Gr√°fica de la relaci√≥n x y y
plt.show()
```

```python
plt.plot(x, y, 'ro') # Gr√°fica roja y con marcador de puntos
```

```python
plt.plot(x, y, 'bs-') # Gr√°fica azul y con marcador de cuadrados y l√≠nea s√≥lida
plt.show()
```

```python
plt.plot(x, y, 'yD:') # Gr√°fica amarilla y con marcador de diamantes y l√≠nea punteada
plt.show()
```

```python
plt.hist(x) # Histograma de x
plt.show()
```

```python
plt.pie(x) # Gr√°fica de pastel de x
plt.show()
```

```python
plt.scatter(x, y) # Gr√°fica de dispersi√≥n de x y y
plt.show()
```

```python
plt.boxplot(x) # Diagrama de caja de x
plt.show()
```

```python
plt.bar(x, y) # Gr√°fica de barras de x y y
plt.show()
```

```python
plt.barh(x, y) # Gr√°fica de barras horizontales de x y y
plt.show()
```

```python
plt.stem(x, y) # Gr√°fica de tallos y hojas de x y y
plt.show()
```
