```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset ("tips")

tips.head()
```

```python
sns.histplot(data=tips, x= 'tip', bins=15, hue='sex')
plt.show()
```

```python
sns.histplot(data=tips, x= 'tip', bins=15, cumulative=True) # cumulative para ver el comportamiento acumulativo de las frecuencias en el histograma
plt.show()
```

```python
sns.histplot(data=tips, x= 'tip', bins=15, hue='sex', stat='frequency') # Con el stat te puede trabajar frecuencia, probabilidad ... ect
plt.show()
```

```python
sns.histplot(data=tips, x= 'tip', bins=15, hue='sex', stat='density') # Con el stat te puede trabajar frecuencia, probabilidad ... ect
plt.show()
```

```python
sns.histplot(data=tips, x= 'tip', bins=15, hue='sex', stat='count', multiple='dodge') # multiple sieve para especifica cómo se deben representar múltiples distribuciones en un solo gráfico
plt.show()
```

```python
sns.kdeplot(data=tips, x= 'tip', hue='sex', shade=True, bw_adjust=1)  # Diagrama de densidad | shade para rellenar el area | bw_adjust para ajustar la suavidad de la curva
plt.show()
```

```python
sns.ecdfplot(data=tips, x= 'tip', hue='sex', stat='count') # Función de distribución acumulativa empírica
```

```python
sns.displot(data=tips, x= 'tip', hue='sex', kind='kde', multiple= 'stack') # Displot es una función que combina histograma y kdeplot
plt.show()
```
