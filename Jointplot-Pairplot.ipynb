```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex') # jointplot te da una combinación de lo que crea que son son los mejore gráficos para representar los datos
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='hist', marginal_ticks=True) # Aunque también lo podemos personalizar con kind | marginal_ticks es para poner las escala del otro histograma
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='hist', marginal_ticks=True,
              marginal_kws=dict(bins=25, fill=False, multiple='dodge')) # Especificaciones del histograma de arriba
plt.show()
```

```python
sns.pairplot(data=tips) # pairplot para relaciones entre variables numéricas
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True) # corner es para no tener información duplicada
plt.show()
```
