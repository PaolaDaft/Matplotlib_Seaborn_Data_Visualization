```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
tips.corr(numeric_only=True) # para sacar la correlaci√≥n en una tabla
```

```python
sns.heatmap(tips.corr(numeric_only=True)) # para la correlaci√≥n 
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True) # annot pone los valores la correlaci√≥n 
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=5, linecolor='black') # cmap para los colores | linewidths para el ancho de la linea | linecolor para el color de la linea
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=5, linecolor='black',
            vmin=0, vmax=1) # ajustamos los valores min y max de la correlaci√≥n
plt.show()
```
