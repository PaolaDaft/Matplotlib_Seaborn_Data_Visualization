```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100) # un arreglo de 100 elementos aleatorios entre 1 y 50
data
```

## Histograma
Muestra la distribuci√≥n de los datos

```python
plt.hist(data, bins=10, alpha=0.5, color='b', edgecolor='black') # Bins es el n√∫mero de barras, alpha es la transparencia, edgecolor es el color de los bordes
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step', edgecolor='magenta') 
plt.show()
```

## Boxplot

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True ) # vert es para que la gr√°fica sea horizontal, patch_artist es para que se rellene el cuadro y notch es para que tenga una muesca (Enfocado en la mediana)
plt.show()
```

```python
data = np.append(data, [100, 200, 300, 400]) # Agregamos 4 valores at√≠picos
plt.boxplot(data, vert=False, patch_artist=True, notch=True ) # vert es para que la gr√°fica sea horizontal, patch_artist es para que se rellene el cuadro y notch es para que tenga una muesca (Enfocado en la mediana)
# Se puede observar que los valores at√≠picos se representan como puntos
# Si agrego el atributo showfliers=False, los valores at√≠picos no se mostrar√°n
plt.show()
```

## Scatterplot
Gr√°fica de dispersion

```python
N = 50
x = np.random.rand(N) # Genera un arreglo de N elementos aleatorios entre 0 y 1
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2
```

```python
plt.scatter(x, y, s=area, c=colors, alpha=0.5) # s es el tama√±o de los puntos, c es el color de los puntos
# El atributo mareker es para cambiar la forma de los puntos (marker='x'), (marker='o'), (marker='^')
plt.show()
```
