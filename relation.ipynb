```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') # style pone el estilo de los puntos de a cuerdo a la var indicada | size es para el no de comensales
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') # style pone el estilo de los puntos de a cuerdo a la var indicada | size es para el no de comensales
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5)) # para mover la leyenda | loc es la ubicacin de la leyenda | bbox_to_anchor es para mover la leyenda
plt.show()
```

```python
plt.figure(figsize=(10, 5)) # para cambiar el tamao de la figura
markers = {'Lunch': 'D', 'Dinner': 's'} # para cambiar los marcadores
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers) 
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5)) 
plt.show()
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') 
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5)) 
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') # relplot es para hacer un scatterplot, relacin con variables
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time') 
plt.show()
```

**pairplot** te permite ver comparativas de pares de variables y analizar su comportamiento en un solo grfico.

```python
sns.pairplot(data=tips,hue='day')
plt.show() 
```
