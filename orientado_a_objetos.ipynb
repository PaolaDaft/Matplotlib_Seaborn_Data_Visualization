```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure() # Este es el lienzo
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # Los primeros dos parámetros son la posición del gráfico en el lienzo, los segundos dos son el tamaño del gráfico
# izquierda, abajo, ancho, alto (rango de 0 a 1)

# Lo que vamos a pintar / gráficas
axes.plot(x, y, 'r')
fig.show()
```

```python
fig = plt.figure() # Este es el lienzo
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # izquierda, abajo, ancho, alto (rango de 0 a 1)
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3]) # Objetos distintos que reposan en un mismo lienzo/ figura

# Lo que vamos a pintar / gráficas
axes.plot(x, y, 'r')
axes2.plot(y, x, 'b')

axes2.set_facecolor('gray')
axes2.set_title('Gráfico pequeño')

fig.show()
```
