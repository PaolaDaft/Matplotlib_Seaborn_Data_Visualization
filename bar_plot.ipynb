```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
countries = ['USA', 'Brazil', 'India', 'Russia', 'South Africa']
population = [331002651, 212559417, 1380004385, 145934462, 59308690]
```

```python
plt.bar(countries, population, color=['pink','salmon'], width=0.5) # width es para el ancho de las barras
plt.xticks(np.arange(5),('Estados Unidos', 'Brasil', 'India', 'Rusia', 'Sudafrica'), rotation=45) # Cambiar las etiquetas del eje x
plt.xlabel('Países')
plt.ylabel('Población')
plt.title('Población de algunos países')
plt.show()
```

```python
countries = ['Perú', 'Argentina', 'Colombia', 'Mexico', 'Brasil']
population = [25.199748, 36.24159, 39.234062, 98.821456, 170.179643]
colors = ['firebrick', 'skyblue', 'gold', 'green', 'limegreen']

plt.style.use("seaborn")
fig, ax = plt.subplots() # fig es la figura y ax es el eje
bars = ax.barh(countries, population, color=colors, label="Cant. habitantes") # Crear las barras

ax.bar_label(bars)
ax.set_xlabel("Cant. de habitantes ($millones$)")
ax.set_title("Países con mas población de Latam")

plt.show()
```
