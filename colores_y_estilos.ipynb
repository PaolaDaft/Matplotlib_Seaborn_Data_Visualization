```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
print(plt.style.available) # Los estilos con los que puedo trabajar por default
```

```python
plt.style.use('dark_background') # Selecciono el estilo que quiero usar
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, 'r')
ax.plot(x, x+2, 'b')
ax.plot(x, x+3, 'y')
ax.plot(x, x+4, 'g')
```

## Modificaciones de tipo MATLAB

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
```

## Modificaciones con PYPLOT

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, color='blue', alpha=0.5) # el alpha es el nivel de transparencia
ax.plot(x, x+2, color='#8B008B', linewidth= 5) # linewidth es el ancho de la linea
ax.plot(x, x+3, color='salmon', linestyle = '--') # linestyle es el estilo de la linea
ax.plot(x, x+4, color='chartreuse', marker = 'P', markersize = 10, markerfacecolor='pink') 
# marker es el marcador, markersize es el tamao del marcador y markerfacecolor es el color del marcador
```

```python
#  Hack para poder ver todos los estilos aplicados
for i in plt.style.available:
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(5,5))
  ax.plot(x, y)
```
