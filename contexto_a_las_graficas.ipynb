```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(x, y, 'b', label='$Sin(x)$')
axes[0].set_title('Relacion X y Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend(loc=0) # Para que muestre la leyenda en la esquina superior izquierda, si no lo pones no se muestra 'label='Sin(x)''

axes[1].plot(y, x, 'r')
axes[1].set_title('Relacion Y y X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')

```

```python
plt.plot(x, y, label = 'Sin(x)')# Sin el mtodo orientado a objetos
plt.title('Relacion X y Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(0.5, 0.2)) # Abajo a la izquierda, con bbox_to_anchor puedes mover la leyenda a donde quieras
```
