```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) # dodge sirve para separar los puntos
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) # En este caso los puntos no se superponen, se ven mejor la concentracin de los datos
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) # dodge sirve para separar los puntos
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, y='total_bill') # El box plot tambin funciona para variables numricas
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) 
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color='0') # el swarmplot se puede combinar con otros grficos
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='pastel')  # Con el violinplot se puede ver la distribucin de los datos | split divide los violines por sexo(los combina)
plt.show()
```

```python
sns.violinplot(data=tips, y='total_bill') 
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='swarm'),  # catplot es una funcin que permite combinar varios grficos para variables categricas
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='box', col='time'),  # Col divide los grficos por la variable indicada
plt.show()
```
